```python
# DataSpelli Notebook — Autochess Strategy Analysis

import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# -----------------------------
# Dosya yolları
# -----------------------------
log_dir = Path("output/strategy_logs")
game_endings_file = log_dir / "game_endings.jsonl"
placement_file = log_dir / "placement_events.jsonl"
market_file = log_dir / "market_events.jsonl"

# -----------------------------
# JSONL dosyalarını yükleme
# -----------------------------
def load_jsonl(file_path):
    data = []
    with open(file_path, "r") as f:
        for line in f:
            data.append(json.loads(line))
    return data

game_endings = load_jsonl(game_endings_file)
placements = load_jsonl(placement_file)
markets = load_jsonl(market_file)

# -----------------------------
# DataFrame oluşturma
# -----------------------------
df_games = pd.DataFrame(game_endings)
df_placements = pd.DataFrame(placements)
df_markets = pd.DataFrame(markets)

# -----------------------------
# TEMPO ANALİZİ
# -----------------------------
tempo_games = df_games[df_games["winner"]=="tempo"]
tempo_placements = df_placements[df_placements["player"]=="tempo"]

# Snowball — average turn for dominance
snowball_turns = tempo_games["turn_count"].mean()
print(f"Tempo average game length: {snowball_turns:.2f} turns")

# Placement ratio (center vs rim)
center_count = tempo_placements["hex_coord"].apply(lambda x: x=="center").sum()
rim_count = tempo_placements["hex_coord"].apply(lambda x: x=="rim").sum()
total_placements = center_count + rim_count
print(f"Tempo placement ratio — center: {center_count/total_placements:.2%}, rim: {rim_count/total_placements:.2%}")

# Combo score dağılımı
plt.figure(figsize=(8,4))
sns.histplot(tempo_placements["combo_score"], bins=20, kde=True, color="red")
plt.title("Tempo Placement Combo Score Distribution")
plt.xlabel("Combo Score")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# STRATEJİ KARŞILAŞTIRMASI
# -----------------------------
# Win rate karşılaştırması
win_counts = df_games["winner"].value_counts()
plt.figure(figsize=(8,4))
sns.barplot(x=win_counts.index, y=win_counts.values, palette="viridis")
plt.title("Strategy Win Counts")
plt.ylabel("Wins")
plt.xlabel("Strategy")
plt.show()

# Center placement oranı strateji bazında
center_placements = df_placements.groupby("player")["hex_coord"].apply(lambda x: (x=="center").sum()/len(x))
plt.figure(figsize=(8,4))
sns.barplot(x=center_placements.index, y=center_placements.values, palette="coolwarm")
plt.title("Center Placement Ratio per Strategy")
plt.ylabel("Center Placement %")
plt.xlabel("Strategy")
plt.show()
```
